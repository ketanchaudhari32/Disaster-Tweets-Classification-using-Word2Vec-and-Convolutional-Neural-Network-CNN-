```
import pandas as pd
from sklearn.model_selection import train_test_split

import nltk
from nltk.tokenize import word_tokenize

import warnings
warnings.filterwarnings('ignore')

nltk.download('punkt') #uncomment o n frst run
```

```
df = pd. read_csv('processedData/Cleaned_Data.csv')

df.dropna(inplace=True)
df.head()
```

```
X = df['Tweet']
Y = df['Label']
#X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1)

#total_vocab=df['Tweet'][df['Label'] == 1]
total_vocab=df['Tweet']
#print([x.strip(' ').split() for x in total_vocab])
```

Creating a Word2Vec model using gensim library

```

import gensim
from gensim.models.word2vec import Word2Vec
```

```
model_sg = Word2Vec(sg=1, size=100, negative=5, window=2, min_count=2, alpha=0.065, min_alpha=0.065)
model_sg.build_vocab([x.strip(' ').split() for x in total_vocab])
#model_sg.train([x.strip(' ').split() for x in X_train], total_examples=model_sg.corpus_count, epochs=model_sg.epochs)
model_sg.train([x.strip(' ').split() for x in X], total_examples=model_sg.corpus_count, epochs=model_sg.epochs)
```

```
#model_cbow = Word2Vec(sg=0,size=200 , min_count=10)
model_cbow = Word2Vec(sg=0, size=100, negative=5, window=2, min_count=2, alpha=0.065, min_alpha=0.065)
#model_cbow.build_vocab([x.strip(' ').split() for x in X_train])
model_cbow.build_vocab([x.strip(' ').split() for x in total_vocab])
#model_cbow.train([x.strip(' ').split() for x in X_train], total_examples=model_cbow.corpus_count, epochs=model_cbow.epochs)
model_cbow.train([x.strip(' ').split() for x in X], total_examples=model_cbow.corpus_count, epochs=model_cbow.epochs)
```

```
model_sg.most_similar('fire')
```

```
model_cbow.most_similar(['fire'])
```

Saving the Word2vec model so as to load it again when required

```
model_sg.save('saved_models/tweets_word2vec_sg.model')
model_cbow.save('saved_models/tweets_word2vec_cbow.model')
```
