```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

```

```python
cleaned_df = pd.read_csv('processedData/Cleaned_Data.csv')

cleaned_df = cleaned_df.dropna()
cleaned_df.head()
```

```python
def plot_words(df):
    all_words = " ".join(df['Tweet'])
    wordcloud = WordCloud(width = 800, height = 600, random_state = 20, max_font_size = 100).generate(all_words)
    plt.figure(figsize = (10,7))
    plt.imshow(wordcloud, interpolation = 'bilinear')
    plt.axis('off')
    plt.show()
```

```python
df_infor = cleaned_df[cleaned_df['Label'] == 1]
plot_words(df_infor)
```

```python
df_NonInfo = cleaned_df[cleaned_df['Label'] == 0]
plot_words(df_NonInfo)
```
